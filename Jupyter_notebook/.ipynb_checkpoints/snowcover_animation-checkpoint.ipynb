```python
%matplotlib inline
import os
import glob
import rasterio
import fiona
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from IPython.display import HTML
import seaborn as sns
```

```python
year = 2016
outdir = "animation"
```

```python
snowcover_dir = "/Volumes/CWMobileSSD/Geodata_fires/2020_COVID_AFS/natice_noaa/"
geotiff_subdir = 'GeoTIFF_Products/'
geotiff_dir = os.path.join(snowcover_dir, geotiff_subdir)
tiffiles = sorted(glob.glob(os.path.join(geotiff_dir, f"*{year}*.tif")))
```

```python
psa_file = "/Volumes/CWMobileSSD/Geodata_fires/AICC/AK_PSA_boundaries/predictive_service_areas.shp"
psa_shapes = {}
with fiona.open(psa_file, "r") as shapefile:
    for feature in shapefile:
        psa_shapes[feature['properties']['PSA_NAME']] = feature["geometry"]
```

```python
sampletiff = tiffiles[65]
f"{sampletiff[90:94]}-{sampletiff[94:96]}-{sampletiff[96:98]}"
```

```python
imgsrc = rasterio.open(sampletiff)
```

```python
plt.imshow(imgsrc.read(1), interpolation='nearest')
```

```python
cmap = mpl.colors.ListedColormap(['white', 'navy', 'orangered', 'lightskyblue', 'ghostwhite'])
cmap.set_over('0.25')
cmap.set_under('0.75')
bounds = list(np.arange(6) - 0.5)
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
pl = ax.imshow(imgsrc.read(1), cmap=cmap, norm=norm, interpolation='nearest')
ax.set_axis_off()
ax.text(250, 250, f"{sampletiff[90:94]}-{sampletiff[94:96]}-{sampletiff[96:98]}", fontsize=14)
```

```python
for sampletiff in tiffiles:
    datestmp = f"{sampletiff[90:94]}-{sampletiff[94:96]}-{sampletiff[96:98]}"
    imgsrc = rasterio.open(sampletiff)
    fig, ax = plt.subplots(figsize=(12, 8))
    pl = ax.imshow(imgsrc.read(1), cmap=cmap, norm=norm, interpolation='nearest')
    ax.set_axis_off()
    ax.text(250, 250, datestmp, fontsize=14)
    plt.savefig(os.path.join(outdir, f"{datestmp}_snow{year}.png"), bbox_inches="tight")
    print(f"written {datestmp}_snow{year}.png")
    fig.clf()
```

```python
plt.rcParams["animation.html"] = "jshtml"
import matplotlib.animation


t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
l, = ax.plot([0,2*np.pi],[-1,1])

animate = lambda i: l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))
ani
```

```python
ani.save('sine_wave.gif', writer='imagemagick')
```

```python
import matplotlib.animation as animation


fps = 30
nSeconds = 5
snapshots = [ np.random.rand(5,5) for _ in range( nSeconds * fps ) ]

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure( figsize=(8,8) )

a = snapshots[0]
im = plt.imshow(a, interpolation='none', aspect='auto', vmin=0, vmax=1)

def animate_func(i):
    if i % fps == 0:
        print( '.', end ='' )

    im.set_array(snapshots[i])
    return [im]

anim = animation.FuncAnimation(
                               fig, 
                               animate_func, 
                               frames = nSeconds * fps,
                               interval = 1000 / fps, # in ms
                               )
```
